```python
import logging
import os
from gensim import corpora, utils
from gensim.models.wrappers.dtmmodel import DtmModel
import numpy as np
import pickle as pkl
# if not os.environ.get('DTM_PATH', None):
#     raise ValueError("SKIP: You need to set the DTM path")
```

```python
with open('cleaned_yearwise.pkl','rb') as fp:
    cleaned_yearwise = pkl.load(fp)
    
```

```python
all_documents = []
total=0
for year,docs in cleaned_yearwise.iteritems():
    all_documents.extend(docs)
    total+=len(docs)
```

```python
class DTMcorpus(corpora.textcorpus.TextCorpus):

    def get_texts(self):
        return self.input

    def __len__(self):
        return len(self.input)

corpus = DTMcorpus(all_documents)
```

```python
len(corpus.dictionary.values())
```

```python
float(total)/len(corpus.dictionary.values())
```

```python
time_seq = [len(val) for key,val in cleaned_yearwise.iteritems()]
```

```python
dtm_path = "/home/avikbasu/WORK/Economics_of_Innovation/TopicEvolution/dtm/dtm/dtm"
model = DtmModel(dtm_path, corpus, time_seq, num_topics=2,
                 id2word=corpus.dictionary, initialize_lda=True)
```

```python
topics = model.show_topic(topicid=1, time=1, num_words=10)
```

```python
with open('dtm_model.pkl','wb') as fp:
    pkl.dump(model, fp)
```
